```python
#Calculating Mean

import pandas as pd 
import numpy as np

sales = [120,85,200,95,150,180,110]

#Method 1
mean_manual = sum(sales) / len(sales)
print('Manual Mean:', mean_manual)

#Method 2
mean_numpy = np.mean(sales)
print('Numpy Mean: ', mean_numpy)

#Method 3
df = pd.DataFrame({'daily_sales': sales})
mean_pandas = df['daily_sales'].mean()
print('Pandas Mean:', mean_pandas)          

```

```python
#Calculating Median


import pandas as pd 
import numpy as np

sales = [120, 85, 5000, 95, 150, 180, 110]

# Method 1 — Manual implementation
sorted_s = sorted(sales)
n = len(sorted_s)
if n % 2 == 1:   # odd
    median_manual = sorted_s[n // 2]
else:             # even
    median_manual = (sorted_s[n//2 - 1] + sorted_s[n//2]) / 2
print(f'Manual Median: {median_manual}')

# Method 2 — NumPy
import numpy as np
print(f'NumPy Median: {np.median(sales)}')

# Method 3 — Pandas
import pandas as pd
df = pd.DataFrame({'sales': sales})
print(f'Pandas Median: {df["sales"].median()}')
```

```python
#Calculating Mode

from scipy import stats
import pandas as pd
from collections import Counter

purchases = [3, 1, 2, 3, 4, 3, 2, 1, 3, 5]

# Method 1 — scipy (handles multimodal correctly)
result = stats.mode(purchases)
print(f'Mode: {result.mode}  |  Count: {result.count}')

# Method 2 — Pandas
df = pd.DataFrame({'purchases': purchases})
print(f'Pandas Mode: {df["purchases"].mode().tolist()}')  # Returns list (handles bimodal)

# Method 3 — Manual using Counter
counts = Counter(purchases)
max_freq = max(counts.values())
modes = [k for k,v in counts.items() if v == max_freq]
print(f'Manual Mode(s): {modes}  (each appears {max_freq} times)')
```

```python
#Calculating Variance

import numpy as np
import pandas as pd

sales = [120, 85, 200, 95, 150, 180, 110]

# NumPy SAMPLE variance — ddof=1 means divide by (n-1)
var_sample = np.var(sales, ddof=1)
print(f'Sample Variance:     {var_sample:.2f}')   # Expected: 1903.54

# NumPy POPULATION variance — ddof=0 (default) means divide by n
var_pop = np.var(sales, ddof=0)
print(f'Population Variance: {var_pop:.2f}')   # Smaller — uses n not n-1

# Pandas — sample variance by default (ddof=1)
df = pd.DataFrame({'sales': sales})
print(f'Pandas Variance:     {df["sales"].var():.2f}')

# CRITICAL: Always specify ddof=1 for sample variance in NumPy.
# ddof = degrees of freedom. Pandas defaults to ddof=1. NumPy defaults to ddof=0.
```

```python
#Calculating Standard Deviation

import numpy as np
import pandas as pd

sales = [120, 85, 200, 95, 150, 180, 110]

# NumPy SAMPLE standard deviation (ddof=1)
std_sample = np.std(sales, ddof=1)
print(f'Sample SD:     {std_sample:.2f}')     # Expected: 43.63

# NumPy POPULATION standard deviation (ddof=0)
std_pop = np.std(sales, ddof=0)
print(f'Population SD: {std_pop:.2f}')

# Pandas (ddof=1 by default — sample SD)
df = pd.DataFrame({'sales': sales})
print(f'Pandas SD:     {df["sales"].std():.2f}')

# Full descriptive summary — one of the most used commands in data science
print(df['sales'].describe())
# Output: count, mean, std, min, 25% (Q1), 50% (median), 75% (Q3), max
```
