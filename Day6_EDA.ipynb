```python
#The Standard Workflow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/Premium/Desktop/Claude Data Management/Titanic/Titanic-Dataset.csv')

print(df.shape)      # rows and columns
print(df.head())     # first 5 rows
print(df.tail())     # last 5 rows
print(df.info())     # column names, types, non-null counts
print(df.describe()) # statistics for numerical columns
```

```python
# How many missing values per column
print(df.isnull().sum())

# As a percentage
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Histogram for every numerical column
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()

# Box plot — shows median, spread, and outliers
df.boxplot(figsize=(12, 6))
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Age'].hist(bins=20, color='steelblue', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
df['Fare'].hist(bins=30, color='coral', edgecolor='black')
plt.title('Distribution of Passenger Fares')
plt.xlabel('Fare (GBP)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
df[['Age', 'Fare']].boxplot()
plt.title('Box Plot — Age and Fare')
plt.show()
```

```python
survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print(survival_by_gender)

survival_by_gender.plot(kind='bar', color=['coral', 'steelblue'], 
                        edgecolor='black')
plt.title('Survival Rate by Gender (%)')
plt.xlabel('Gender')
plt.ylabel('Survival Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_by_class)

survival_by_class.plot(kind='bar', color=['gold', 'steelblue', 'coral'],
                       edgecolor='black')
plt.title('Survival Rate by Passenger Class (%)')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f')
plt.title('Correlation Matrix — Titanic')
plt.show()
```
